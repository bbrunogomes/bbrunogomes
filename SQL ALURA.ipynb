## Ciência de dados

Com o Colab, você pode aproveitar todo o potencial das conhecidas bibliotecas Python para analisar e ver dados. A célula de códigos abaixo usa <strong>numpy</strong> para gerar dados aleatórios e <strong>matplotlib</strong> para visualizá-los. Para editar o código, basta clicar na célula e começar a editar.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

3. Importar as Bibliotecas Necessárias:
Primeiro, importamos as bibliotecas que serão usadas no código. **sqlite3** é utilizada para interagir com o banco
de dados SQLite e **pandas** é utilizada para manipular dados em
formato de DataFrame.


```
import sqlite3
import pandas as pd
```

4. Conectar-se ao Banco de Dados SQLite:
O comando **sqlite3.connect('dados.db')** cria uma conexão com o banco de dados SQLite chamado **'dados.db'**. Se o banco de dados ainda não existir, ele será criado. A conexão é atribuída à variável conn, que será usada para executar operações no banco de dados.

```
# Conecta-se ao banco de dados SQLite (ou cria um novo se não existir)
conn_fornecedores = sqlite3.connect('dados_forncederores.db')
conn_pedidos = sqlite3.connect('dados_pedidos.db')
```

5. Carregar os Dados do Arquivo CSV para um DataFrame do Pandas:
Utilizamos a função **pd.read_csv()** do Pandas para ler os dados do arquivo CSV **'tabela+fornecedores (1).csv'** e carregá-los em um DataFrame chamado dados.

```
# Carrega os dados do arquivo CSV para um DataFrame do Pandas
dados_fornecedores = pd.read_csv('tabela+fornecedores (1).csv')
dados_pedidos = pd.read_csv('tabela+pedidos.csv')
```

6. Salvar o DataFrame no Banco de Dados SQLite:
Utilizamos o método **to_sql()** do Pandas para salvar os dados do DataFrame dados na tabela **'tabela_fornecedores'** no banco de dados SQLite. O parâmetro **if_exists='replace'** indica que se a tabela já existir no banco de dados, ela será substituída pelos novos dados. O parâmetro index=False especifica que não queremos salvar o índice do DataFrame como uma coluna no banco de dados.




```
dados_fornecedores.to_sql('tabela_fornecedores', conn_fornecedores, if_exists='replace', index=False)
dados_pedidos.to_sql('tabela_pedidos', conn_pedidos, if_exists='replace', index=False)
```

7. Fechar a Conexão com o Banco de Dados:
Por fim, fechamos a conexão com o banco de dados SQLite usando o método close() da variável conn. Isso é importante para liberar os recursos e garantir que os dados sejam salvos corretamente no banco de dados.

```
conn_fornecedores.close()
conn_pedidos.close()
```

### SELECT
O comando **SELECT** é uma das instruções mais básicas em SQL e é usado para recuperar dados de um banco de dados.




```
# Conectar-se ao banco de dados SQLite
# Selecione tudo da tabela tabela_fornecedores
%sql sqlite:///dados_forncederores.db
%sql SELECT * FROM tabela_fornecedores LIMIT 3
```

### WHERE
A cláusula WHERE é usada para filtrar os resultados de uma consulta com base em uma condição especificada. Ela permite que você extraia apenas os registros que atendem a certos critérios.

```
# Selecione tudo na tabela tabela_fornecedores onde o Páis de Origem é igual a China
# No google colabe usamos o nome "País de Origem"
# Em uma IDE SQL usamos o nome da tabela ex: pais_de_origem para filtrar
%sql SELECT * FROM tabela_fornecedores WHERE "País de Origem" = 'China';
```


---

### DISTINCT
A palavra-chave DISTINCT é usada para retornar apenas valores distintos em uma consulta. Ela elimina valores duplicados de uma coluna ou conjunto de colunas do resultado da consulta.

```
%sql sqlite:///dados_pedidos.db
%sql SELECT * FROM tabela_pedidos LIMIT 3
```

```
# Selecione apenas valores unicos da coluna cliente da tabela_pedidos
%sql SELECT DISTINCT cliente FROM tabela_pedidos
```

### CHAVE ESTRANVEIRA

Uma chave estrangeira (foreign key) em um banco de dados relacional é um campo ou um conjunto de campos que estabelece uma relação entre duas tabelas.

 Responsável pelo relacionamento entre tabelas.

 Utilizados campos que são unicos como id_categoria

 image.png

Você está desenvolvendo um banco de dados para um supermercado online. Para isso, você cria uma tabela para os produtos em estoque e outra para os fornecedores. O objetivo é estabelecer uma relação entre essas duas tabelas usando chaves estrangeiras. Você cria a tabela de produtos com a seguinte instrução:

```
%sql
CREATE TABLE produtos (
    id INT PRIMARY KEY,
    nome VARCHAR(255),
    descricao TEXT,
    categoria INT,
    preco DECIMAL(10, 2),
    unidade VARCHAR(50),
    fornecedor INT,
    data_inclusao DATE
);
```

Agora você precisa incluir a chave estrangeira na tabela de produtos para relacioná-la com o fornecedor do produto, através do campo id da tabela fornecedores.

```
%sql
ALTER TABLE produtos
%sql ADD FOREIGN KEY (fornecedor) REFERENCES fornecedores(id)
```
